```python
import matplotlib.pyplot as plt
from pyts.image import RecurrencePlot

# Redefining the sliding windows function
def sliding_windows(data, window_size, step_size):
    windows = [
        data.iloc[i:i+window_size].values
        for i in range(0, data.shape[0] - window_size + 1, step_size)
    ]
    return np.array(windows)

# Specifying window size and step size for the sliding windows
window_size = 30
step_size = 1

# Select sensor data for unit number 2 and specifically for sensor measurement 2
unit_2_data = train_df[train_df['unit_number'] == 2]["sensor_measurement_2"]

# Create sliding window sequences for sensor measurement 2 data
sequence_data_unit_2 = sliding_windows(unit_2_data, window_size, step_size)

# Generating recurrence plots for the sequences
rp_transformer = RecurrencePlot(threshold='point', percentage=20)  # Using 20% distance threshold
rp_images = rp_transformer.fit_transform(sequence_data_unit_2)

# Plotting all the recurrence plot images
num_plots = len(rp_images)  # This determines how many subplots we need
cols = 5  # Define number of columns in the plot grid
rows = num_plots // cols + (num_plots % cols > 0)  # Calculate required rows

plt.figure(figsize=(15, rows * 3))  # Adjusting figure size appropriately
for i in range(num_plots):
    ax = plt.subplot(rows, cols, i + 1)
    ax.imshow(rp_images[i], cmap='binary', origin='lower')
    cycle_start = i * step_size + 1
    ax.set_title(f'Cycle {cycle_start}')  # Title showing the start of each cycle for the window
    ax.axis('off')  # Hide axes
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# 데이터 로드
data_path =  'C:/Users/Administrator/RUL2/cmapss/train_FD001.txt'
data = pd.read_csv(data_path, sep=" ", header=None)
data.dropna(axis=1, how='all', inplace=True)

# 열 이름 지정
column_names = ['unit_nr', 'time_cycles', 'setting1', 'setting2', 'setting3'] + [f'sensor_{i}' for i in range(1, 22)]
data.columns = column_names
```

```python
# 최대 사이클 수 계산
data['max_cycles'] = data.groupby('unit_nr')['time_cycles'].transform(max)

# RUL 계산
data['RUL'] = data['max_cycles'] - data['time_cycles']
```

```python
def generate_rp_matrix(data, threshold=0.1):
    n = data.shape[0]
    rp_matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            distance = np.linalg.norm(data.iloc[i] - data.iloc[j])
            rp_matrix[i, j] = 1 if distance < threshold else 0
    return rp_matrix
```

```python
from pyts.image import RecurrencePlot
import numpy as np

# 시계열 데이터 예시
data_length = 100
window_size = 30
time_series = np.random.rand(data_length)

# 슬라이딩 윈도우 적용 및 RP 계산
rp_transformer = RecurrencePlot(threshold=None)
num_windows = data_length - window_size + 1
rps = []

for start in range(num_windows):
    end = start + window_size
    window = time_series[start:end].reshape(1, -1)
    rp = rp_transformer.fit_transform(window)
    rps.append(rp[0])

# 첫 번째 RP의 Shape 출력
print("Shape of the first Recurrence Plot:", rps[0].shape)
```
